```python
# MNIST   

import numpy as np # NumPy   Python Import
import tensorflow as tf # tensorflow library   Python Import
import tensorflow.keras.datasets as ds # ds   tensorflow.keras       Python Import
import matplotlib.pyplot as plt # Matplotlib  , plt    


from tensorflow.keras.models import Sequential # TensorFlow Keras  Sequential    
from tensorflow.keras.layers import Dense # TensorFlow Keras  Dense    
from tensorflow.keras.optimizers import SGD # TensorFlow Keras  SGD(Stochastic Gradient Descent)    


# MNIST   
(x_train, y_train), (x_test, y_test) = ds.mnist.load_data() # MNIST   



#  
x_train = x_train.reshape(60000, 784).astype(np.float32) / 255.0 # MNIST    `x_train`  
x_test = x_test.reshape(10000, 784).astype(np.float32) / 255.0 # MNIST     `x_test`  
y_train = tf.keras.utils.to_categorical(y_train, 10) # MNIST     `y_train` - (one-hot encoding)   
y_test = tf.keras.utils.to_categorical(y_test, 10) # MNIST     `y_test` - (one-hot encoding)   



# Create the model( )
mlp = Sequential() # Sequential    
mlp.add(Dense(units=512, activation='tanh', input_shape=(784,))) # Sequential  Dense   
mlp.add(Dense(units=10, activation='softmax')) # Sequential    Dense   



# Compile the model( )
mlp.compile(loss='categorical_crossentropy', optimizer=SGD(learning_rate=0.01), metrics=['accuracy']) # Sequential   
#       
history = mlp.fit(x_train, y_train, batch_size=128, epochs=50, validation_data=(x_test, y_test), verbose=2) # Sequential   
#   (  )
plt.figure(figsize=(12, 4)) #    



#     
plt.subplot(1, 2, 1) # 1 2       
plt.plot(history.history['accuracy']) #Matplotlib     
plt.plot(history.history['val_accuracy']) #Matplotlib     
plt.title('Model Accuracy') #Matplotlib      
plt.xlabel('Epoch') #Matplotlib   x   
plt.ylabel('Accuracy') #Matplotlib   y   
plt.legend(['Train', 'Validation'], loc='upper left') # Matplotlib    (legend)  



#      
plt.subplot(1, 2, 2) #Matplotlib      (subplot) ,      
plt.plot(history.history['loss']) #Matplotlib   (loss)     
plt.plot(history.history['val_loss']) #Matplotlib    (validation loss)     
plt.title('Model Loss') #Matplotlib    (title)  
plt.xlabel('Epoch') #Matplotlib   x (label)  
plt.ylabel('Loss') #Matplotlib   y (label)  
plt.legend(['Train', 'Validation'], loc='upper left') #Matplotlib    (legend)  



plt.show() #Matplotlib     



#    
res = mlp.evaluate(x_test, y_test, verbose=0) #    
print('Accuracy=', res[1] * 100) #     

```
